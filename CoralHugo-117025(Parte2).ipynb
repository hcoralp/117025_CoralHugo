```python
'''
Punto 16.

El programa lee una lista de números ya ordenados de forma ascendente y 
verifica que dicha lista está ordenada. Luego, lee un número y lo inserta 
en la lista en la posición que le corresponde a dicho número.
'''

lista=eval(input("Ingrese una lista de números: "))
i=0
j=1
k=0

while j < len(lista): 
    if lista[i]==lista[j]: # Busca que no se repitan los números.
        print("En la lista hay números repetidos.")
        j=1 
        break
    elif lista[i]<lista[j]: # Verifica que los números se encuentren
        i+=1                # ordenados en forma ascendente.
        j+=1
    elif lista[i]>lista[j]: 
        print("La lista no está ordenada de forma ascendente.")
        k=1
        break

if k==0: # Si los números están de forma ascendente, este continúa.
    print("La lista se encuentra ordenada de forma ascendente.")
    
    x=float(input("Ingrese un número: "))
    z=0 
    while z < len(lista):
        if x==lista[z]: # Verifica si el número ya está en la lista.
            print("El número ya se encuentra en la lista.")
            z=1
            break
        elif x > lista[z]: # Busca la pocisión en la que debería estar 
            k+=1           # el número.
        elif x < lista[z]: # Posiciona el número cuando encuentra uno menor.
            lista.insert(z,x)
            break
        if x > lista[-1]: # Dado que el número ingresado sea mayor
                          # a todos en la lista, se agrega al final.
            lista.append(x)
            break
            
    if k== 0 : print("La nueva lista es: ",lista)
```

```python
'''
Punto 17.

El programa lee una lista y determinar cuál es el 
segundo elemento más grande de dicha lista.
'''

x=eval(input("Ingrese una lista: " ))

i=0 
MAY=x[0] # Pone como elemento mayor al primero de la lista.

while i<len(x):
    if MAY<x[i]: # Compara el elemento mayor hasta el momento con el 
                 # siguiente en la lista, si ese elemto es mayor, entonces 
                 # actualiza la variable 'MAY'.
        MAY=x[i]
        i+=1
    else: i+=1
    
while MAY in x: # Elimina el elemento mayor de la lista 'L' las veces que 
                # se encuentre.
    x.remove(MAY)

t=0
MAY1=x[0]
while t<len(x): # Vuelve a realizar el mismo procedimiento para buscar el 
                # mayor de la lista resultante, que es el segundo mayor de la 
                # lista ingresada.
    if MAY1<x[t]:
        MAY1=x[t]
        t+=1
    else: t+=1 

print("El segundo elemento más grande de la lista es: ",MAY1)
```

```python
'''
Punto 18.

Leer una matriz cuadrada A, obtener su matriz triangular superior 
asociada U y determinar qué fila de U contiene la mayor cantidad de unos.
'''

matriz = []
orden=int(input("Matriz de orden: ")) # Se define el orden de la matriz.

for i in range(orden):
    matriz.append([0]*orden)
    
for f in range(orden):     # Aquí se define los elementos
    for c in range(orden): # y la posición que van a tener.
        matriz[f][c]=int(input("Elemento %d,%d : " % (f,c)))

print("La matriz que ingresó es: ")
print( )
for filas in matriz:
    print(filas)

for i in range(len(matriz)):
    for j in range(len(matriz[i])):
        if i-j > 0:
            matriz[i][j]=0

print("La matriz triangular asociada es: ")
print( )
for filas in matriz:
    print(filas)
```

```python
'''
Punto 19.

El programa lee una lista e identifique la sublista más larga dentro de 
dicha lista que se encuentra ordenada ascendentemente.
'''

x=eval(input('Ingrese una lista de números: '))

U=[] #crea dos listas que se irán actualizando cuando encuentren númmeros 
     # organizados
V=[]
i=0
t=1
while t<len(x)-1:
    if  x[i]<x[t]:    # Agrega elementos a la lista U, cuando los 
                      # encuentra ordenados.
        U.append(x[i])
        i+=1
        t+=1
    
    if x[i]>x[t] and len(U)>len(V): # Si la lista 'U' es más grande que la 
                                    # lista 'V', entonces actualiza 'V'.
        U.append(x[i])
        V=U[:]
        U=[]   # Vacía la lista 'U' para que sigan entrando valores.
        if t<len(x)-1: 
            i+=1
            t+=1
            
    if x[i]>x[t] and len(U)<len(V): # Si encuentra una discontinuidad y la 
                                    # lista U es menor a la lista 'V', 
                                    # entonces descarta la lista 'U'.
        U=[]
        if t<len(x)-1:
            i+=1
            t+=1
    
    if x[i]<x[t] and t==len(x)-1: # Considera si los últimos dígitos deben
                                  # ser agregados a la lista 'U'.
        U.append(x[i])
        U.append(x[t])
        if len(U)>len(V): # Compara cúal es la mayor.
            
            print("La sublista ordenada más grante es: ",U)
        else:
            print("La sublista ordenada más grande es: ",V)
```

```python
'''
Punto 20.

El programa lee una matriz y la rota 90 grados 
en sentido antihorario.
'''

n=int(input("Cantidad de filas: "))
m=int(input("Cantidad de columnas: "))

M1=[] # Crea una lista vacía que almacenará los datos de la matriz.
for fila in range(n): # Se crea una matriz de ceros con la cantidad de
                      # filas y columnas indicadas.
    M1.append([0]*m)

fila=0
col=0
while fila < n:
    while col < m: # Realiza el ingreso de los valores de la matriz.
        x=int(input(f"Ingrese el valor de la posición {fila+1},{col+1}: "))
        M1[fila][col]=x
        col+=1
    fila+=1
    col=0

print("Su matriz es: ")
print( )
for fila in M1:
    print(fila)
    

M2=[] # Crea una lista vacía que almacenará los datos de la matriz.
for fila in range(m): # Se crea una matriz de ceros con la cantidad 
                      # de filas y columnas indicadas.
    M2.append([0]*n)


# Variables que indican el índice para cambiar los valores.

fil1=0 # Indica la posición de fila para M1.
fil2=m-1 # Indica la posición fila de M2.
col1=0 # Posición columna de M1.
col2=0 # Posición columna de M2.
while col2 >= 0: # Condiciona el ciclo infinitamente.
    if fil1 == n or col1 == m: # Rompe el ciclo cuando se ha recorrido
                               # toda la matriz ingresada.
        break
    while fil2 >= 0:
        M2[fil2][col2]=M1[fil1][col1] # Cambia los valores de M2 en su 
                                      # posición correspondiente.
        fil2-=1
        col1+=1

    fil1+=1
    col2+=1
    col1=0
    fil2=m-1 # Reinicia la posición de 'fil2'.
            
print("La matriz rotada 90° en sentido antihorario es: ")
print( )

for fila in M2:
    print(fila)
```

```python
'''
Punto 20a.

El programa lee una matriz y la rota 90 grados 
en sentido antihorario. Utilizando el módulo numpy.
'''

import numpy as np

matriz = []
filas=int(input("Cantidad de filas: "))        # Aquí definimos el tamaño
columnas=int(input("Cantidad de columnas: "))  # que va a tenar la matriz.

for i in range(filas):
    matriz.append([0]*columnas)
    
for f in range(filas):          # Con esto definimos los elementos
    for c in range(columnas):   # y que posición van a tener.
        matriz[f][c]=int(input("Elemento %d,%d : " % (f,c)))

print("Su matriz es: ")
print( )
for filas in matriz:
    print(filas)

mat=np.rot90(matriz) # Haciendo uso del módulo numpy se rota
                     # la matriz 90° en sentido antihorario.
print("La matriz rotada 90° en sentido antihorario es: ")
print( )
for filas in mat:
    print(filas)
```
